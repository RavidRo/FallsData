```python
import pandas as pd
import numpy as np
```

```python
data_directory = "Falls_one_class_100_balanced"
raw_data = pd.read_csv(data_directory + "/raw_data.csv")
vmap = pd.read_csv(data_directory + "/vmap.csv")
raw_data
```

```python
raw_data
```

```python
temporal_propertiy_ids = raw_data["TemporalPropertyID"].unique()
temporal_propertiy_ids
```

```python
properties_timestamps = {}
properties_values = {}
for temporal_property_id in temporal_propertiy_ids:
    times = raw_data[raw_data["TemporalPropertyID"] == temporal_property_id]["TimeStamp"].values
    values = raw_data[raw_data["TemporalPropertyID"] == temporal_property_id]["TemporalPropertyValue"].values
    properties_timestamps[temporal_property_id] = times
    properties_values[temporal_property_id] = values
properties_timestamps
```

```python
properties_time_values = {}
for property in properties_timestamps.keys():
    times_values_pairs = {}
    times = properties_timestamps[property]
    values = properties_values[property]
    for i in range(len(times)):
        if(times[i] not in times_values_pairs):
            times_values_pairs[times[i]] = [values[i]]
        else:
            times_values_pairs[times[i]].append(values[i])
    for i in range(len(times)):
        times_values_pairs[times[i]] = np.mean(times_values_pairs[times[i]]) # mean of all entities

    properties_time_values[property] = times_values_pairs
properties_time_values
```

```python
trend_size = 3
property_trends = {}
for property in properties_time_values.keys():
    trends = []
    times_values_pairs = properties_time_values[property]
    for i in range(trend_size + 1, len(times_values_pairs.keys())):
        x = np.array(list(map(lambda idx: list(times_values_pairs.keys())[idx], range(i-trend_size-1, i))))
        y = np.array(list(map(lambda timestamp: times_values_pairs[timestamp], x)))
        nb_points = np.size(y)
        mean_x, mean_y = np.mean(x), np.mean(y)
        a = np.sum(x * y)
        b = mean_x * sum(y)
        c = mean_x * mean_y
        ss_xy = np.sum(x * y) - mean_x * sum(y) - mean_y * sum(x) + nb_points * mean_x * mean_y
        ss_xx = np.sum(x * x) - 2 * mean_x * np.sum(x) + nb_points * mean_x * mean_x
        if ss_xx != 0:
            b_1 = ss_xy / ss_xx
            trends.append(np.degrees(np.arctan(b_1)))
    property_trends[property] = trends
property_trends
```

```python
c = 0
for property in property_trends:
    n = len(property_trends[property])
    sorted_trends = sorted(property_trends[property])
    decreasing_prop = int(0.45 * n)
    increasing_prop = int(0.45 * n)
    decreasing_trends = sorted_trends[:decreasing_prop]
    increasing_trends = sorted_trends[-increasing_prop:]
    stable_trends = sorted_trends[decreasing_prop:-increasing_prop]
    # if(property == 55):

    # print(f"property: {property}")
    # print(f"all trends: {len(sorted_trends)}")
    # print(f"increasing: {(increasing_trends)}")
    # print(f"decreasing: {(decreasing_trends)}")
    # print(f"stable: {(stable_trends)}")
    # print(f"stable: {stable_trends}")
    # print(f"decreaing bin: [{min(decreasing_trends)},{max(decreasing_trends)}]")
    # print(f"stable bin: [{min(stable_trends)},{max(stable_trends)}]")
    # print(f"increasing bin: [{min(increasing_trends)},{max(increasing_trends)}]")
    if (min(decreasing_trends) == max(decreasing_trends)) or (min(stable_trends) == max(stable_trends)) or (min(increasing_trends) == max(increasing_trends)):
        print(property)
```

```python
gradient_file_df = pd.DataFrame()
bin_id_counter = 0
state_counter = 1
for property in property_trends:
    n = len(property_trends[property])
    sorted_trends = sorted(property_trends[property])
    decreasing_prop = int(0.45 * n)
    increasing_prop = int(0.45 * n)
    # decreasing_trends = sorted_trends[:decreasing_prop]
    increasing_trends = sorted_trends[-increasing_prop:]
    # stable_trends = sorted_trends[decreasing_prop:-increasing_prop]
   
    bin_low_stable = min(stable_trends)
    bin_high_stable = max(stable_trends)
    bin_high_decreasing = bin_low_stable
    bin_low_decreasing = min(min(decreasing_trends), bin_high_decreasing - 1)
    bin_low_increasing = bin_high_stable
    bin_high_increasing = min(max(increasing_trends), bin_low_increasing + 1)
   

    dict = {'StateID': state_counter, 'TemporalPropertyID': property, 'Method': 'gradient', 'BinID': bin_id_counter % 3, 'BinLow': bin_low_decreasing, 'BinHigh': bin_high_decreasing}
    gradient_file_df = gradient_file_df.append(dict, ignore_index = True)
    bin_id_counter += 1
    state_counter += 1

    dict = {'StateID': state_counter, 'TemporalPropertyID': property, 'Method': 'gradient', 'BinID': bin_id_counter % 3, 'BinLow': bin_low_stable, 'BinHigh': bin_high_stable}
    gradient_file_df = gradient_file_df.append(dict, ignore_index = True)
    bin_id_counter += 1
    state_counter += 1

    dict = {'StateID': state_counter, 'TemporalPropertyID': property, 'Method': 'gradient', 'BinID': bin_id_counter % 3, 'BinLow': bin_low_increasing, 'BinHigh': bin_high_increasing}
    gradient_file_df = gradient_file_df.append(dict, ignore_index = True)
    bin_id_counter += 1
    state_counter += 1

gradient_file_df = gradient_file_df.astype({'StateID':'int', 'TemporalPropertyID':'int', 'BinID':'int'})
gradient_file_df.to_csv(data_directory + '\\gradient.csv', index=False)
gradient_file_df
```
